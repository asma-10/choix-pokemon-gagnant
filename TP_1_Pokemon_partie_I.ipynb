## L'univers des Pokémons

Un Pokémon est un animal issu du monde des jeux vidéo. Chaque Pokémon possède des caractéristiques qui lui sont propres, à savoir son type (Pokémon d'herbe, Pokémon de feu ... ) lui donnant alors des facultés spéciales. Ainsi, le Pokémon nommé `Dracofeu`, de type `Feu` est capable de cracher du feu. De plus, on note l'existence de Pokémons légendaires très rares et très puissants (Pokémon diamant, Pokémon platine...).

Les Pokémons sont également de différentes générations faisant référence à l'évolution de l'univers des Pokémons. Les Pokémons de première génération sont ceux ayant pris naissance au tout début de la création de l'univers (création de 151 Pokémons) pour arriver à la huitième génération avec plus de 800 Pokémons (2020)!

Dans cet univers viennent ensuite les dresseurs de Pokémons. Les dresseurs les collectionnent et les élèvent dans le but de les faire combattre entre eux jusqu'au KO. Lors d'un combat de Pokémons, opposant deux dresseurs, l'un choisit l'animal qui sera le premier à entrer dans l'arène. En fonction de ce premier Pokémon, le second dresseur choisit un animal dans sa collection qui sera en mesure de battre le premier Pokémon. Si le premier Pokémon présent dans l'arène est de type feu, il y a de fortes chances que le Pokémon choisi par l'adversaire soit de type eau afin de contrer les attaques du premier.

Cela nécessite pour le dresseur d'avoir une bonne connaissance de chaque Pokémon pour choisir celui qui pourra devenir vainqueur du combat. Pour l'aider dans son choix, les caractéristiques des Pokémons sont répertoriées dans un document appelé Pokédex.

# Les étapes à réaliser pour mener à bien un projet de Machine Learning

1. Définition du problème à résoudre

## Agrégation des données avec le Pokédex

Cette agrégation va nous permettre d'avoir une vue centralisée des données et de permettre une analyse plus approfondie.

#### - Créer une liste comportant :
- Les numéros des Pokémons
- Le nombre de combats menés
- Le nombre de combats gagnés
- Le pourcentage de victoires par rapport au nombre de combats menés

#### - Créer un nouveau Pokedex comportant les noms de Pokemons et leur victoire


```python
combat_df.head()
```

```python
# liste des numéros des Pokémons
pokemone_number_list = np.array(pokemone_df.NUMERO)
```

```python
len(combat_df)
```

```python
# liste de nombre de combats menés
from collections import defaultdict

battles_count_first = combat_df['First_pokemon'].value_counts()
battles_count_second = combat_df['Second_pokemon'].value_counts()

battles_count_addition = battles_count_first.add(battles_count_second, fill_value=0).astype(int)
battles_count = [[index, value] for index, value in battles_count_addition.items()]
battles_count_index = [index for index, value in battles_count_addition.items()]
battles_count_only = [value for index, value in battles_count_addition.items()]
```

```python
list_name = ['NUMERO_pokemon','battles_fought']
battles_count_df = pd.DataFrame(battles_count, columns=list_name)
battles_count_df
```

```python
# liste de nombre de combats gagnés

winned_battles_count_first_pokemon = combat_df[combat_df['First_pokemon'] == combat_df['Winner']]['First_pokemon'].value_counts()
winned_battles_count_second_pokemon = combat_df[combat_df['Second_pokemon'] == combat_df['Winner']]['Second_pokemon'].value_counts()

combine = winned_battles_count_first_pokemon.add(winned_battles_count_second_pokemon, fill_value=0).astype(int)
winned_battles_count = [[index, value] for index, value in combine.items()]
winned_battles_count_index = [index for index, value in combine.items()]
winned_battles_count_only = [value for index, value in combine.items()]
```

```python
list_name = ['NUMERO_pokemon','winned_battles']
winned_battles_count_df = pd.DataFrame(winned_battles_count, columns=list_name)
winned_battles_count_df
```

```python
# Le pourcentage de victoires par rapport au nombre de combats menés
result = []
winnes_vs_battles_fought_percentage = []
for i in range(len(winned_battles_count)):
    if winned_battles_count_index[i] == winned_battles_count_index[i]:
        per = (winned_battles_count_only[i] * 100)/ battles_count_only[i]
        winnes_vs_battles_fought_percentage.append([winned_battles_count_index[i],per])
        # result.append((winned_battles_count[i])[0])
```

```python
list_name = ['NUMERO_pokemon','win_percentage']
winnes_vs_battles_fought_percentage_df = pd.DataFrame(winnes_vs_battles_fought_percentage, columns=list_name)
winnes_vs_battles_fought_percentage_df
```

```python
pokemone_num = list(set(combat_df['First_pokemon'].tolist() + combat_df['Second_pokemon'].tolist()))
```

```python
pokemone_new_df = pd.merge(battles_count_df, winned_battles_count_df, on='NUMERO_pokemon')
```

```python
# Le nouveau Pokedex

New_pokedex = pd.merge(pokemone_new_df, winnes_vs_battles_fought_percentage_df, on='NUMERO_pokemon')
```

```python
New_pokedex = New_pokedex.rename(columns={'NUMERO_pokemon' : 'NUMERO'})
```
